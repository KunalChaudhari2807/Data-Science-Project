```python
plt.figure()
plt.scatter(df.sepal_length,df.sepal_width,alpha=0.9,color='Orange')
plt.title('ires spirsl')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.hist(df['sepal_length'],alpha=0.9,color='#8C3582',bins=20,label='Alt Yaprak Uzunluk')
plt.hist(df['sepal_width'],alpha=0.9,color='#10AD9D',bins=20,label='st Yaprak Uzunluk')
plt.title('Alt Yaprak Uzunluu ve st Yaprak Uzunluu - Histogram')
plt.xlabel('CM')
plt.ylabel('Frekans')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(df['sepal_length'],alpha=0.5,label='Alt Yaprak Uzunluk',color='r')
plt.xlabel('CM')
plt.ylabel('Freakans')
plt.legend()
plt.show()
```

```python
fig, ax=plt.subplots(2, 1)
fig.suptitle('Alt Yaprak Uzunluk ve st Yaprak Uzunluk',fontsize=14)
ax[0].plot(df.sepal_length,color='g',label='SepalLengthCm')
ax[0].set_title('Alt Yaprak Uzunluk')
ax[0].set_xlabel('sepal')
ax[0].set_ylabel('CM')

ax[1].plot(df.petal_length,color='r',label='PetalLengthCm')
ax[1].set_title('st Yaprak Uzunluk')
ax[1].set_xlabel('petal')
ax[1].set_ylabel('CM')

fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show()
```

```python
species_counts = df['species'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(species_counts, labels=species_counts.index, colors=['pink', 'yellow', 'red'], autopct='%1.1f%%')
plt.show()
```
